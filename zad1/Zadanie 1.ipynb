## Zadanie 1  - Porównanie rozkładów pośpieszności $y$ i pseudopośpieszności $\eta$

Badane cząstki:

$\pi^0$, masa $m_{\pi^0} = 134.98 MeV/c^2$

$K^0$, masa $m_{K^0} = 497.65 MeV/c^2$

$D^0$, masa $m_{D^0} = 1864.84 MeV/c^2$

Algorytm programu:
1. Losuj n razy $y$ i $p_T$
2. Oblicz $p_z = m_T \sinh y$, gdzie $m_T^2 = m^2 + p_T^2$
3. Oblicz $\cos\theta = \frac{p_z}{p} = \frac{p_z}{\sqrt{p_T^2 + p_z^2}}$
4. Oblicz pseudopośpieszność $\tanh\eta = \cos\theta \rightarrow \eta$ = arctanh($\cos\theta$)
5. Rozkłady dla $y$ i $\eta$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Losujemy y i p_T z przedziału 0-1, rozkład jednorodny

# n - lizba wylosowanych liczb
n = 500000
y_random = [np.random.uniform(-5, 5) for _ in range(n)]
p_T_random = [np.random.uniform(0, 2000) for _ in range(n)]
```

```python
# Liczymy m_T

# Masy w MeV
m_pi = 134.98
m_K = 497.65
m_D = 1864.84

p_z_pi = [np.sqrt((m_pi**2 + p_T**2)) * np.sinh(y) for p_T, y in zip(p_T_random, y_random)]
p_z_K = [np.sqrt((m_K**2 + p_T**2)) * np.sinh(y) for p_T, y in zip(p_T_random, y_random)]
p_z_D = [np.sqrt((m_D**2 + p_T**2)) * np.sinh(y) for p_T, y in zip(p_T_random, y_random)]
```

```python
cos_theta_pi = [p_z / np.sqrt(p_T**2 + p_z**2) for p_z, p_T in zip(p_z_pi, p_T_random)]
cos_theta_K = [p_z / np.sqrt(p_T**2 + p_z**2) for p_z, p_T in zip(p_z_K, p_T_random)]
cos_theta_D = [p_z / np.sqrt(p_T**2 + p_z**2) for p_z, p_T in zip(p_z_D, p_T_random)]
```

```python
eta_pi = [np.arctanh(cos_theta) for cos_theta in cos_theta_pi]
eta_K = [np.arctanh(cos_theta) for cos_theta in cos_theta_K]
eta_D = [np.arctanh(cos_theta) for cos_theta in cos_theta_D]
```

```python
# Parametry do wykresów

# X axis range
x_min = -15
x_max = 15

# number of bins
n_bins = 200

# figsize
f_size = (14, 9)
```

```python
plt.figure(figsize=f_size)
y_values, _, _ = plt.hist(eta_pi, bins=n_bins, range=(x_min, x_max), histtype='step')

plt.title('Rozkład $\eta$ dla $\pi^0$', fontsize=18)
plt.xlabel('$\eta$', fontsize=15)
plt.ylabel('Zliczenia', fontsize=15)
plt.text(9, np.max(y_values) * 0.75, f'Total counts: {n} \n# bins: {n_bins} \nMean: {np.mean(eta_pi):.4f} \nStd dev: {np.std(eta_pi):.4f}', fontsize=16)
# plt.grid(alpha=0.7, axis='y')

plt.savefig('rozklad_pi.png')
plt.show()
```

```python
plt.figure(figsize=f_size)
y_values, _, _ = plt.hist(eta_K, bins=n_bins, range=(x_min, x_max), histtype='step')

plt.title('Rozkład $\eta$ dla $K^0$', fontsize=18)
plt.xlabel('$\eta$', fontsize=15)
plt.ylabel('Zliczenia', fontsize=15)
plt.text(9, np.max(y_values) * 0.75, f'Total counts: {n} \n# bins: {n_bins} \nMean: {np.mean(eta_K):.4f} \nStd dev: {np.std(eta_K):.4f}', fontsize=16)
# plt.grid(alpha=0.7, axis='y')

plt.savefig('rozklad_K.png')
plt.show()
```

```python
plt.figure(figsize=f_size)
y_values, _, _ = plt.hist(eta_D, bins=n_bins, range=(x_min, x_max), histtype='step')

plt.title('Rozkład $\eta$ dla $D^0$', fontsize=18)
plt.xlabel('$\eta$', fontsize=15)
plt.ylabel('Zliczenia', fontsize=15)
plt.text(9, np.max(y_values) * 0.75, f'Total counts: {n} \n# bins: {n_bins} \nMean: {np.mean(eta_D):.4f} \nStd dev: {np.std(eta_D):.4f}', fontsize=16)
# plt.grid(alpha=0.7, axis='y')

plt.savefig('rozklad_D.png')
plt.show()
```

```python
plt.figure(figsize=f_size)
y_values, _, _ = plt.hist(y_random, bins=n_bins, range=(-1, 1), histtype='step')

plt.title('Rozkład $y$', fontsize=18)
plt.xlabel('$y$', fontsize=15)
plt.ylabel('Zliczenia', fontsize=15)
plt.text(-.1, np.max(y_values) * 0.2, f'Total counts: {n} \n# bins: {n_bins} \nMean: {np.mean(y_random):.4f} \nStd dev: {np.std(y_random):.4f}', fontsize=16)
# plt.grid(alpha=0.7, axis='y')

plt.savefig('rozklad_y.png')
plt.show()
```
